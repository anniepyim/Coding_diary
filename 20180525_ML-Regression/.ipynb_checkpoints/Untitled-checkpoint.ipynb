```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

from scipy import stats
from scipy.stats import norm, skew #for some statistics

%matplotlib inline
#set color
sns.set_palette("Set2")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
#correlation map
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(df.corr(),annot=True, linewidths=.1, fmt= '.1f', ax=ax)
```

# Data Processing
The output variable is sales price. Let's check the distribution of the variable

```python
sns.distplot(df['SalePrice'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(df['SalePrice'], plot=plt)
plt.show()
```

## Data Transformation
Since the data is positively skewed, lets make normalize it.

### For positively skewed data, options are:
- Square or cube root (mild)
- Logarithmic (substential)

### For negatively skewed data, options are:
- Square or root, or even higher orders

```python
fig,((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3, figsize=(15,10))
sns.distplot(np.sqrt(df["SalePrice"]) , fit=norm, ax=ax1);
ax1.set(title='Sqrt transformation')

res = stats.probplot(np.sqrt(df["SalePrice"]), plot=ax4)
ax4.set(title='Sqrt transformation Q-Q')

sns.distplot(np.cbrt(df["SalePrice"]) , fit=norm, ax=ax2);
ax2.set(title='Cbrt transformation')

res = stats.probplot(np.cbrt(df["SalePrice"]), plot=ax5)
ax5.set(title='Cbrt transformation Q-Q')

sns.distplot(np.log1p(df["SalePrice"]) , fit=norm, ax=ax3);
ax3.set(title='Log transformation')

res = stats.probplot(np.log1p(df["SalePrice"]), plot=ax6)
ax6.set(title='Log transformation Q-Q')

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df['SalePrice'])
ax3.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
```

Looks like Log transformation is the best.

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
df["SalePrice"] = np.log1p(df["SalePrice"])
```

## Missing Data

Missing data could sometimes have disastrous effect on your analysis. For continous variables, you could choose to fill it with the mean, or median, or impute by some methods. Now let's be simple and just fill with median.

Make sure to do this for the test data too!

```python
#filling NA's with the mean of the column:
df = df.fillna(df.median())
```

For categorical ones, you could choose to fill with the cat that occured most frequently. Let's do it later
