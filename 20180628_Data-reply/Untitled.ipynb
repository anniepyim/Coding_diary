```python
def connect(x,y):
    result = False
    if (x == "A") & (y == "B"):
        result = True
    if (x == "B") * (y == "C"):
        result = True
    return result
```

```python
ex = ["A", "B", "C", "D"]

group_num = 0
k=1
group = {}

for x in ex:
    for y in ex[k:]:
        print(x,y, connect(x,y))
        if (connect(x,y) == True):
            if (x not in group) & (y not in group):
                group[x] = group_num
                group[y] = group_num
                group_num += 1
            elif (x in group) & (y in group):
                if group[x] < group[y]:
                    group[y] = group[x]
                else:
                    group[x] = group[y]
            elif (x in group):
                group[y] = group[x]
            else:
                group[x] = group[y]
        else:
            if (x not in group):
                group[x] = group_num
                group_num += 1
            if (y not in group):
                group[y] = group_num
                group_num += 1
    k += 1
    
inv_group = {}

for k,v in group.items():
    if v in inv_group:
        inv_group[v].append(k)
    else:
        inv_group[v] = [k]

result = []
for k,v in inv_group.items():
    result.append(v)
    
print(result)
```
