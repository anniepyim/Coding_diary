```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

#set color
sns.set_palette("Set2")
```

```python
#plt.figure returns a Figure object
fig = plt.figure()

#fig.add_subplot returns an Axes object (where we can draw stuff) that belongs to fig
#arg = row, col, pos of the axis
ax1 = fig.add_subplot(221)
ax1.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax1.set(xlim=[0.5, 4.5], ylim=[10, 30], title='An Example Axes',
       ylabel='Y-Axis', xlabel='X-Axis')

ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

plt.show()
```

```python
#plt.subplot returns Figure object and (a set of) Axes object that belong to that fig
#It is equivalent to the first method, but simpler way of declaration of objects

#fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)

fig, axes = plt.subplots(2,2)
ax1, ax2, ax3, ax4 = axes.flatten()

ax1.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax1.set(xlim=[0.5, 4.5], ylim=[10,30], title='An Example Axes',
       ylabel='Y-Axis', xlabel='X-Axis')

plt.show()
```

```python
# plt.plot makes plot and act on Axes object of that plot implicitly
# Act on current Axes object
# Useful when make simple individual dirty plots
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)

#this actually calls ax.set_xlim()
plt.xlim([0.5, 4.5])
plt.show()
```

```python
iris = pd.read_csv("data/Iris.csv")

iris.head()
```

```python
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
# This function combines regplot() and FacetGrid. 
# It is intended as a convenient interface to fit regression models across conditional subsets of a dataset.
sns.lmplot( x="SepalLengthCm", y="SepalWidthCm", data=iris, fit_reg=True, hue='Species', legend=False)
plt.legend(loc='lower right')
```

```python
# Combine regplot() and PairGrid when kind="reg
sns.pairplot(iris.drop(['Id'], axis=1), hue="Species",kind="reg")
```

```python
#Taking only two of the variables
sns.pairplot(x_vars=['SepalLengthCm'], y_vars=['SepalWidthCm'], data=iris, hue="Species", size=5)
```

The plotting functions in seaborn are broadly divided into two classes

- "Axes-level" functions, including:
    - stripplot, swarmplot, boxplot, violinplot, barplot, pointplot, countplot (categorical)
    - regplot, residplot (regression)
    - kdeplot, distplot (distribution)
- "Figure-level" functions, including:
    - lmplot (combine regplot and FacetGrid (ie could have more than one plot, each of diff factor according to your variable))
    - factorplot (combine categorical plot (e.g. kind="violin") with FacetGrid)
    - jointplot (use JointGrid (i.e. plot two variables with two types of graphs)
    - pairplot (use PairGrid)

The first group is identified by taking an explicit ax argument and returning an Axes object. As this suggests, you can use them in an "object oriented" style by passing your Axes to them

Axes-level functions will only draw onto an Axes and won't otherwise mess with the figure, so they can coexist perfectly happily in an object-oriented matplotlib script.

The second group of functions (Figure-level) are distinguished by the fact that the resulting plot can potentially include several Axes which are always organized in a "meaningful" way. That means that the functions need to have total control over the figure, so it isn't possible to plot, say, an lmplot onto one that already exists. Calling the function always initializes a figure and sets it up for the specific plot it's drawing

https://stackoverflow.com/questions/23969619/plotting-with-seaborn-using-the-matplotlib-object-oriented-interface

```python
# Plots such as violin plot and box plot could be plot on an axis
fig, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(2,3, figsize=(18, 10))
sns.violinplot(x="Species", y="SepalLengthCm", data=iris, ax=ax1)
sns.boxplot(x="Species", y="SepalWidthCm", data=iris, ax=ax2)
sns.swarmplot(x="Species",y="PetalLengthCm", data=iris, ax=ax3)
sns.barplot(x="Species", y="PetalWidthCm", data=iris, ax=ax4)

sns.countplot(x="Species",data=iris, ax=ax5)
sns.kdeplot(data=iris["PetalLengthCm"], data2=iris["PetalWidthCm"], ax=ax6)

# could even overlap
sns.stripplot(x="Species", y="SepalWidthCm", data=iris, jitter=True, ax=ax2)
```
